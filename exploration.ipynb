1) Qu’est-ce que l’apprentissage automatique supervisé ?
L’apprentissage automatique supervisé est une méthode où un modèle est entraîné à partir de données étiquetées, c’est-à-dire des exemples pour lesquels on connaît la réponse. Le but est d’apprendre à prédire une valeur cible (target) en se basant sur des caractéristiques d’entrée (features). Par exemple, prédire si un client va souscrire à une offre bancaire en fonction de son âge, son métier, etc.

2) Qu’est-ce que les données étiquetées (ou labellisées) ?
Les données labellisées sont des données pour lesquelles on connaît déjà la valeur de sortie attendue. Chaque observation contient donc à la fois des entrées (features) et une étiquette cible (label). Exemple :
[âge = 35, emploi = technicien, ...] → y = "oui" (le client a souscrit).

3) Qu’est-ce que la classification supervisée ?
C’est une tâche d’apprentissage supervisé où l’objectif est de prédire une classe ou une catégorie. Dans notre projet, il s'agit de prédire si un client va souscrire ("yes") ou non ("no") à un dépôt à terme, donc une classification binaire.

4) Qu’est-ce qu’un modèle d’apprentissage automatique ?
C’est un programme ou une fonction qui apprend des relations entre les données d’entrée et la cible. Une fois entraîné, ce modèle peut faire des prédictions sur de nouvelles données qu’il n’a jamais vues.

5) Qu’est-ce que les données d’entraînement ? Et qu’est-ce que l’entraînement d’un modèle de classification supervisée ?
Les données d’entraînement sont un sous-ensemble des données utilisées pour apprendre au modèle comment faire des prédictions.
L’entraînement consiste à ajuster les paramètres internes du modèle à partir de ces données, pour qu’il apprenne à prédire correctement la cible.

6) Qu’est-ce que la donnée cible (ou "target") ?
C’est la variable à prédire. Dans le dataset "Bank Marketing", la variable cible est la colonne y, qui indique si un client a souscrit (yes) ou non (no) à un dépôt à terme.

7) À quoi sert la phase d’entraînement d’un modèle d’apprentissage automatique ?
Cette phase sert à faire apprendre au modèle les liens entre les variables d’entrée et la cible. Il découvre des patterns dans les données pour pouvoir faire des prédictions fiables plus tard.

8) À quoi sert la phase de prédiction d’un modèle d’apprentissage automatique ?
Une fois entraîné, le modèle est utilisé pour prédire la cible sur de nouvelles données. C’est la phase où on exploite ce que le modèle a appris.

9) Qu’est-ce que le prétraitement des données et pourquoi est-il important ?
C’est l’ensemble des étapes pour préparer les données avant de les donner au modèle :

Nettoyage (valeurs manquantes)

Transformation (encodage des variables catégorielles)

Normalisation ou standardisation
Sans prétraitement, le modèle risque de mal fonctionner ou d’apprendre des choses biaisées.

10) Comment évaluer un modèle de classification avec la métrique "accuracy" ?
L’accuracy (ou taux de précision) mesure le pourcentage de prédictions correctes.
Formule :
accuracy = (prédictions correctes / total des prédictions)
Exemple : si le modèle prédit correctement 90 clients sur 100, l'accuracy est de 90 %.
⚠️ Attention : en cas de déséquilibre des classes (ex : 90% de "no", 10% de "yes"), l’accuracy peut être trompeuse.

11) Qu’est-ce que l’AutoML et pourquoi c’est utile pour un débutant ?
L’AutoML (Automated Machine Learning) automatise les étapes complexes du machine learning :

Prétraitement

Choix et test de modèles

Optimisation des hyperparamètres
C’est parfait pour les débutants car ça permet de créer des modèles efficaces sans avoir besoin de tout coder à la main. PyCaret est un bon exemple d’outil AutoML.

12) Qu’est-ce qu’un arbre de décision pour la classification supervisée ? Comment fonctionne-t-il ?
C’est un modèle de classification qui fonctionne comme un arbre de questions.
Chaque nœud de l’arbre représente une condition sur une variable (ex : "âge > 30 ?").
À chaque étape, on "branche" à gauche ou à droite en fonction de la réponse, jusqu’à arriver à une prédiction finale (feuille de l’arbre).
